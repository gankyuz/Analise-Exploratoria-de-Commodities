```python
import pandas as pd
import numpy as np
```

Carregando conjunto de dados

```python
df = pd.read_csv("COMMODITY_DATA.csv")
```

Explorando conjunto

```python
df = df.replace(',', '', regex=True)
df = df.replace('-', '', regex=True)
df = df.replace('', np.nan)
df = df.replace('MAY90', np.nan)

df= df.dropna()
```

```python
df.isnull().sum()
```

```python
df = df.rename(columns={'WTI': 'PETROLEO WTI'})
df = df.rename(columns={'COTTON': 'ALGODÃO'})
df = df.rename(columns={'NATURAL_GAS': 'GAS NATURAL'})
df = df.rename(columns={'COFFEE': 'CAFÉ'})
df = df.rename(columns={'SUGAR': 'AÇUCAR'})
df = df.rename(columns={'ALUMINUM': 'ALUMÍNIO'})
df = df.rename(columns={'BRENT': 'PETROLEO BRENT'})
df = df.rename(columns={'CORN': 'MILHO'})
df = df.rename(columns={'COPPER': 'COBRE'})
df = df.rename(columns={'WHEAT': 'TRIGO'})
df = df.rename(columns={'date': 'DATA'})
```

```python
lista = [
    'PETROLEO WTI',
    'PETROLEO BRENT',
    'GAS NATURAL',
    'ALGODÃO',
    'CAFÉ',
    'AÇUCAR',
    'ALUMÍNIO',
    'MILHO',
    'COBRE',
    'TRIGO',
    'ALL_COMMODITIES'
]
df[lista].astype(float)
```

```python
df.dtypes
```

```python
df['DATA'] = pd.to_datetime(df['DATA'], format='%Y%m%d')
df = df.set_index('DATA')
```

Análise Exploratória das fontes energia (Petróleo e Gás Natural).

```python
from ast import increment_lineno
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline

sns.set_style("whitegrid")
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

Correlação entres as fontes de energia

```python
lst = ['PETROLEO WTI', 'PETROLEO BRENT', 'GAS NATURAL']
corrmat = df[lst].corr()

fig = plt.figure(figsize=(12, 9))
mask = np.triu(np.ones_like(corrmat, dtype=bool))
sns.heatmap(corrmat, vmax=0.8, mask=mask, square=True, annot=True)
plt.show()
```

O mapa de calor demonstra a correlação de preços entre as fontes de energia. Quanto maior valor, maior correlação.

E esse mapa específico aponta uma correlação muito grande entre os Petróleos WTI e Brent.

Qual a variação de preço entre as fontes de energia?

```python
for i in range(len(lst)):
  plt.figure(figsize=(12,12))
  df[lst[i]].hist(figsize=(11,9))
  plt.xlabel('Preço')
  plt.ylabel('Count')
  plt.title(lst[i])
  plt.legend(lst[i], loc='upper center', bbox_to_anchor=(1.2,1))
```

É possível observar que a variação de preços das fontes de energia:
* Petróleo WTI: 40-100 usd.
* Petróleo Brent: 40-120 usd.
* Gás Natural: 2-6 usd.

Variação de preço ao longo dos anos:

```python
plt.figure(figsize=(10,10))
plt.plot(df[['PETROLEO WTI', 'PETROLEO BRENT', 'GAS NATURAL']])
plt.legend(['PETROLEO WTI', 'PETROLEO BRENT', 'GAS NATURAL'], loc='upper center', bbox_to_anchor=(1.2,1))
plt.xlabel('Data')
plt.ylabel('Preço')
```

É possível observar que ao longo dos anos o gás natural manteve valor estável, tendo poucas altas e quedas.

Já os petróleos teve bastante oscilação. Pode se destacar o crescimento a partir dos anos 2000, tendo um queda em 2008, provavelmente relacionado a crise financeira mundial de 2008. Retornou a crescer, mas 2012-2013 teve outra queda, possivelmente ocasionada pela guerra da Síria. E outra queda brusca em 2020, que correlaciono com a pandemia do COVID-19.

Qual o menor preço que os petróleos obtiveram ao longo dos anos?

```python
lst_pet =['PETROLEO WTI', 'PETROLEO BRENT']
plt.figure(figsize=(10,10))
for i in range(len(lst_pet)):
  plt.subplot(2,1, i+1)
  plt.subplots_adjust(hspace= 1, wspace=0.5)
  plt.plot(df[lst_pet[i]])
  plt.xticks(rotation=90)
plt.suptitle('Menor Preço dos Petróleos')
```

O menor preço dos petróleos foi entre os anos 2020 e 2024, tendo um valor abaixo de $25 usd.

Comparação entre categorias de produtos

```python
df_long = df.reset_index().melt(
    id_vars= 'DATA',
    var_name= 'PRODUTO',
    value_name= 'PREÇO'
)
df_long.head()
```

```python
df_long['PRODUTO'].unique()
```

```python

category_map= {
    'PETROLEO WTI': 'ENERGIA',
    'PETROLEO BRENT': 'ENERGIA',
    'GAS NATURAL': 'ENERGIA',
    'ALGODÃO': 'AGRICOLA',
    'CAFÉ': 'AGRICOLA',
    'AÇUCAR': 'AGRICOLA',
    'ALUMÍNIO': 'METAL',
    'MILHO': 'AGRICOLA',
    'COBRE': 'METAL',
    'TRIGO': 'AGRICOLA',
}

df_long['CATEGORIA'] = df_long['PRODUTO'].map(category_map)
```

```python
df_long.tail()
```

```python
print(df_long.groupby('CATEGORIA')['PREÇO'].mean())
```

```python
df_cat = (
    df_long
    .groupby(['CATEGORIA', 'DATA'])['PREÇO']
    .mean()
    .reset_index()
)
```

```python
sns.lineplot(
    data= df_cat,
    x= 'DATA',
    y= 'PREÇO',
    hue= 'CATEGORIA'
)

plt.title('Evolução da média de preços por categoria ao longo dos anos')
plt.show()
```

```python
df_cat_mean = (
    df_long
    .groupby(['CATEGORIA', 'DATA'])['PREÇO']
    .mean()
    .reset_index()
)

plt.figure()
plt.bar(df_cat_mean['CATEGORIA'],df_cat_mean['PREÇO'] )
plt.xlabel('Categoria')
plt.ylabel('Preço Médio')
plt.title('Evolução da média de preços por categoria ao longo dos anos')
```

É possível observar que a categoria metal se destaca em média de valores do que outras.

Análise exploratória entre metais

```python
lst_metal = ['ALUMÍNIO', 'COBRE']

plt.figure(figsize=(10,10))
plt.plot(df[['ALUMÍNIO', 'COBRE']])
plt.xlabel('Anos')
plt.ylabel('Preço')
plt.legend(['ALUMÍNIO', 'COBRE'], loc='upper center', bbox_to_anchor=(1.2,1))
```

Comparação entre produtos agrícolas

```python
plt.figure(figsize=(10,10))

plt.plot(df[[
    'ALGODÃO',
    'CAFÉ',
    'AÇUCAR',
    'MILHO',
    'TRIGO']])

plt.xlabel('Anos')
plt.ylabel('Preço')
plt.legend(['ALGODÃO',
    'CAFÉ',
    'AÇUCAR',
    'MILHO',
    'TRIGO'], loc= 'upper center',bbox_to_anchor=(1.2,1))
```

Observa-se, que entre os produtos de agricultura, o trigo, milho e café possuem valores mais elevados da categoria. No decorrer dos anos tiveram quedas, passou por um tempo com valores baixos, entre 2012-2023, onde teve pico de subida.

```python
lst_agricutural = [  'ALGODÃO',
    'CAFÉ',
    'AÇUCAR',
    'MILHO',
    'TRIGO']

corrmatagr = df[lst_agricutural].corr()
fig_lst = plt.figure(figsize=(12, 9))
mask_list = np.triu(np.ones_like(corrmatagr, dtype=bool))
sns.heatmap(corrmatagr, vmax=0.8, mask=mask_list, square=True, annot=True)
plt.show()
```

O mapa de calor demonstra a correlação de preços produtos de agricultura. Quanto maior valor, maior correlação.

E pode-se concluir, que se tem uma quantidade até positiva de correlação entre esses produtos. Considerando que valor a partir de 0.7 é considerado alto.

Gráfico de comparação entre todos os produtos de todas as categorias

```python
plt.figure(figsize=(10,10))

plt.plot(df[['PETROLEO WTI',
    'PETROLEO BRENT',
    'GAS NATURAL',
    'ALGODÃO',
    'CAFÉ',
    'AÇUCAR',
    'ALUMÍNIO',
    'MILHO',
    'COBRE',
    'TRIGO']])

plt.xlabel('Anos')
plt.ylabel('Preço')
plt.legend(['PETROLEO WTI',
  'PETROLEO BRENT',
  'GAS NATURAL',
  'ALGODÃO',
  'CAFÉ',
  'AÇUCAR',
  'ALUMÍNIO',
  'MILHO',
  'COBRE',
  'TRIGO'], loc= 'upper center',bbox_to_anchor=(1.2,1))
```

De todos os produtos, os metais, cobre e alumínio, obtiveram maiores valores de mercado. Entre os menores preços, milho e gás natural.
